```python
import multiprocessing
import time

def tarea_intensiva(numero):
    resultado = 0
    for _ in range(10**7):  # Simula una tarea computacionalmente intensiva
        resultado += numero
    print(f"Tarea {numero} completada. Resultado: {resultado}")
    return resultado

if __name__ == "__main__":
    numeros = list(range(8))  # Lista de tareas

    inicio = time.time()

    with multiprocessing.Pool(processes=multiprocessing.cpu_count()) as pool:
        resultados = pool.map(tarea_intensiva, numeros)

    fin = time.time()

    print(f"Resultados: {resultados}")
    print(f"Tiempo total: {fin - inicio:.2f} segundos")
```

```python
import concurrent.futures
import time

#... (la función tarea_intensiva es la misma que antes)

if __name__ == "__main__":
    numeros = list(range(8))

    inicio = time.time()

    with concurrent.futures.ProcessPoolExecutor(max_workers=multiprocessing.cpu_count()) as executor:
        resultados = list(executor.map(tarea_intensiva, numeros))

    fin = time.time()

    print(f"Resultados: {resultados}")
    print(f"Tiempo total: {fin - inicio:.2f} segundos")
```

```python
import multiprocessing
import time
import random

def tarea_ciclo(nombre, duracion):
    """Simula una tarea con un ciclo de duración variable."""
    print(f"Tarea {nombre}: Iniciando ciclo de {duracion} segundos.")
    inicio = time.time()
    while time.time() - inicio < duracion:
        # Simula trabajo dentro del ciclo
        _ = random.random() * random.random()
    print(f"Tarea {nombre}: Ciclo completado.")

if __name__ == "__main__":
    tareas = [
        ("A", 100),  # Tarea A: ciclo de 3 segundos
        ("B", 500),  # Tarea B: ciclo de 5 segundos
        ("C", 200),  # Tarea C: ciclo de 2 segundos
        ("D", 400),  # Tarea D: ciclo de 4 segundos
    ]

    procesos = []
    for nombre, duracion in tareas:
        proceso = multiprocessing.Process(target=tarea_ciclo, args=(nombre, duracion))
        procesos.append(proceso)
        proceso.start()

    # Espera a que todos los procesos terminen
    for proceso in procesos:
        proceso.join()

    print("Todas las tareas han finalizado.")
```

```python

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Datos de expresión génica de ejemplo (filas: genes, columnas: muestras)
expresion_genica = np.random.rand(50, 20)

#Crear nombres ficticios para las muestras y los genes.
gene_names = ["gene_"+str(i) for i in range(50)]
sample_names = ["sample_"+str(i) for i in range(20)]

#Convertir a un data frame de pandas para darle los nombres a las filas y columnas.
df = pd.DataFrame(expresion_genica, index = gene_names, columns = sample_names)

# Crear el mapa de calor
plt.figure(figsize=(10, 15)) #tamaño del grafico.
sns.heatmap(df, cmap="viridis") #Creacion del heatmap, y seleccion del esquema de colores.
plt.title("Mapa de calor de expresión génica")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Datos de expresión génica de ejemplo (filas: muestras, columnas: genes)
expresion_genica = np.random.rand(100, 1000)

# Aplicar PCA
pca = PCA(n_components=2)
representacion_2d = pca.fit_transform(expresion_genica)

# Visualizar los datos en 2D
plt.scatter(representacion_2d[:, 0], representacion_2d[:, 1])
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.title("Representación 2D de datos de expresión génica (PCA)")
plt.show()
```
